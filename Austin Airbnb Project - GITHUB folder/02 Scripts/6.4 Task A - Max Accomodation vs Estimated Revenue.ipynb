### 4. Data Prep for Regression Analysis

#### Create a scatterplot using matplotlib for another look at how the chosen variables plot against each other.


```python
df.plot(x = 'accommodates', y='estimated_revenue',style='o') # The style option creates a scatterplot; without it, we only have lines.
plt.title('Listing Max Accomodation vs Estimated Revenue')  
plt.xlabel('Max Accomodation')  
plt.ylabel('Estimated Revenue')  
plt.show()

# will rerun this with review scores vs est_rev
# also number of reviews vs est rev
```

```python
# Reshape the variables into NumPy arrays and put them into separate objects.

X = df['accommodates'].values.reshape(-1,1)
y = df['estimated_revenue'].values.reshape(-1,1)
```

```python
X
```

```python
y
```

```python
# Spliting data into a training set and a test set.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

### 5. Regression analysis

```python
# Creating a regression object.

regression = LinearRegression()

# This is the regression object, which will be fit onto the training set.
```

```python
# Fit the regression object onto the training set.

regression.fit(X_train, y_train)
```

```python
# Predict the values of y using X.

y_predicted = regression.predict(X_test)
```

*Creating a plot that shows the regression line from the model on the test set.*

```python
plot_test = plt
plot_test.scatter(X_test, y_test, color='gray', s = 15)
plot_test.plot(X_test, y_predicted, color='orange', linewidth =3)
plot_test.title('Listed max accomodation vs Estimated Revenue (Test set)')
plot_test.xlabel('Listed max accomodation')
plot_test.ylabel('Estimated Revenue')
plot_test.show()
```

```python
# Create objects that contain the model summary statistics.

rmse = mean_squared_error(y_test, y_predicted) # This is the mean squared error
r2 = r2_score(y_test, y_predicted) # This is the R2 score. 
```

```python
# Print the model summary statistics. This is where you evaluate the performance of the model.

print('Slope:' ,regression.coef_)
print('Mean squared error: ', rmse)
print('R2 score: ', r2)
```

##### With a rmse over 3,000,000 and a R2 Score just over 0.2 (1 being a perfect fit), it's safe to say the model does not fit very well.
