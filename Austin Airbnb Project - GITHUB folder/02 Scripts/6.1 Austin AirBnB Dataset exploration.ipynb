### Bathroom text -> bathrooms variable

```python
df['bathrooms_text'].value_counts(dropna = False)
```

```python
# Split the 'bathrooms_text' column into two separate columns
df[['Bathroom_Numeric', 'Bathroom_String']] = df['bathrooms_text'].str.split('bath', expand=True)
# Remove leading and trailing whitespace from both columns
df['Bathroom_Numeric'] = df['Bathroom_Numeric'].str.strip()
df['Bathroom_String'] = df['Bathroom_String'].str.strip()
```

```python
df['Bathroom_Numeric'].value_counts(dropna= False)
```

```python
df['room_type'].value_counts(dropna = False)
```

```python
df['Bathroom_Numeric'].value_counts(dropna = False)
```

```python
df['Bathroom_Numeric'] = df['Bathroom_Numeric'].replace(['Half-'], '0.5')
```

```python
df_entirespace = df[df['room_type'] =='Entire home/apt']
```

```python
df_sharedspace = df[df['room_type'] !='Entire home/apt']
```

```python
df_entirespace[['Bathroom_Numeric', 'Bathroom_String']] = df['bathrooms_text'].str.split('bath', expand=True)
# Remove leading and trailing whitespace from both columns
df['Bathroom_Numeric'] = df['Bathroom_Numeric'].str.strip()
df['Bathroom_String'] = df['Bathroom_String'].str.strip()
```
